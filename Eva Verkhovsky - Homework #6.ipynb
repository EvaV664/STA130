The theoretical Simple Linear Regression model describes the relationship between a predictor variable (xi) and an outcome variable (yi) by fitting a straight line to the data. The model is represented by the equation: yi=beta0 + beta1xi + error.
- beta0 is the intercept, representing the expected value of y when x is zero.
- beta1 is the slope, indicating the expected change in y for each one-unit increase in x.
- error is the error term, representing the deviation of each observed value from the line. The errors are assumed to be normally distributed with a mean of zero, suggesting that the actual values of y at each x are normally distributed around the regression line.
This structure allows us to estimate the expected values of y given x and understand the underlying relationship between the variables.

```python
import numpy as np
import plotly.graph_objs as go
from scipy.stats import norm, uniform

# Parameters for the simulation
n = 100  # Number of data points
beta0 = 2  # Intercept
beta1 = 3  # Slope
sigma = 1  # Standard deviation of the error term

# Generate predictor variable x from a uniform distribution
x = uniform.rvs(loc=0, scale=10, size=n)

# Generate error term from a normal distribution with mean 0 and standard deviation sigma
errors = norm.rvs(loc=0, scale=sigma, size=n)

# Create the outcome variable y using the Simple Linear Regression model equation
y = beta0 + beta1 * x + errors

# Define the theoretical line without errors (perfectly linear)
y_line = beta0 + beta1 * x

# Plot using plotly
fig = go.Figure()

# Add scatter plot for data points (y values with noise)
fig.add_trace(go.Scatter(x=x, y=y, mode='markers', name='Observed data'))

# Add line plot for the theoretical line (without noise)
fig.add_trace(go.Scatter(x=x, y=y_line, mode='lines', name='Theoretical line (y =  + x)'))

# Update layout for clarity
fig.update_layout(
    title="Theoretical Simple Linear Regression Model",
    xaxis_title="Predictor (x)",
    yaxis_title="Outcome (y)",
    legend_title="Legend"
)

# Show the plot
fig.show(renderer="png")
```

```python
import numpy as np
import pandas as pd
import plotly.express as px
import statsmodels.formula.api as smf

# Parameters for simulation
n = 100  # Number of data points
beta0 = 2  # Intercept
beta1 = 3  # Slope
sigma = 1  # Standard deviation of the error term

# Generate predictor variable x and error term
x = np.random.uniform(0, 10, n)
errors = np.random.normal(0, sigma, n)

# Create outcome variable y based on the linear model
y = beta0 + beta1 * x + errors

# Create a DataFrame with the simulated data
df = pd.DataFrame({'x': x, 'y': y})

# Fit a Simple Linear Regression model using statsmodels
model = smf.ols('y ~ x', data=df).fit()

# Print the model summary
print(model.summary())

# Add predictions to the DataFrame
df['fitted_y'] = model.fittedvalues

# Plot the original data points and the fitted line using plotly.express
fig = px.scatter(df, x='x', y='y', title="Fitted Simple Linear Regression Model with Simulated Data")
fig.add_scatter(x=df['x'], y=df['fitted_y'], mode='lines', name='Fitted Line')

# Show the plot
fig.show(renderer="png")
```

The Ordinary Least Squares (OLS) method is a statistical technique used to fit a linear model to observed data. Here's a concise explanation of what it means:

- Fitted Line: In OLS, the "fitted line" is the straight line that best represents the relationship between the independent (predictor) and dependent (response) variables. This line is determined by minimizing the differences between the observed data points and the predicted values from the line.

- Minimizing Errors: OLS specifically focuses on minimizing the "squared differences" (or errors) between the observed values and the values predicted by the model. The reason for using squares is that it ensures that both positive and negative errors contribute positively to the total error. Squaring the differences emphasizes larger errors and prevents the cancellation of positive and negative errors, leading to a more stable solution.

- Why "Squares"?: By minimizing the sum of the squared errors, OLS provides the best linear approximation of the data in terms of minimizing variance, which leads to the most accurate predictions based on the observed data.

Errors vs residuals 
- Residuals: You calculate these after fitting your model. They help you assess how well the model fits your data.
- Errors: These represent a theoretical concept since the true underlying model is typically unknown. Errors are what would exist if you could compare the actual values directly to the true model values.

    - The denominator measures the total sum of squares (TSS), which represents the total variation in y.
    - By subtracting the ratio of the RSS to the TSS from 1, we obtain the proportion of the total variation in y that is explained by the model. This can be interpreted as the proportion of variation in  yexplained by the fitted model, indicating how well the model fits the data.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Define long wait time threshold and filter the data
long_wait_limit = 63  # you can adjust this as needed
long_wait = old_faithful.waiting >= long_wait_limit
long_wait_data = old_faithful[long_wait]

# Step 1: Bootstrap samples and collect slope coefficients
n_bootstrap_samples = 1000
slope_coefficients = []

for _ in range(n_bootstrap_samples):
    sample = long_wait_data.sample(n=len(long_wait_data), replace=True)
    model = smf.ols('duration ~ waiting', data=sample).fit()
    slope_coefficients.append(model.params['waiting'])

# Convert slope coefficients to a DataFrame for visualization
slope_coefficients_df = pd.DataFrame(slope_coefficients, columns=['slope'])

# Plot the bootstrapped sampling distribution of slope coefficients
plt.figure(figsize=(10, 6))
sns.histplot(slope_coefficients_df['slope'], bins=30, kde=True)
plt.title('Bootstrapped Sampling Distribution of Slope Coefficients')
plt.xlabel('Slope Coefficient')
plt.ylabel('Frequency')
plt.axvline(x=0, color='red', linestyle='--', label='Null Hypothesis (slope = 0)')
plt.legend()
plt.show()

# Step 2: Simulate samples from the regression model under the null hypothesis
beta0 = 1.65
beta1 = 0
sigma = 0.37
n_simulations = 1000
simulated_slopes = []

# Values of waiting (x)
waiting_values = np.linspace(long_wait_data['waiting'].min(), long_wait_data['waiting'].max(), n_simulations)
for _ in range(n_simulations):
    # Simulate y values based on the regression model
    y_simulated = beta0 + beta1 * waiting_values + np.random.normal(0, sigma, waiting_values.shape)
    # Fit model to simulated data
    model_sim = smf.ols('y ~ x', data=pd.DataFrame({'y': y_simulated, 'x': waiting_values})).fit()
    simulated_slopes.append(model_sim.params['x'])

# Convert simulated slopes to a DataFrame for visualization
simulated_slopes_df = pd.DataFrame(simulated_slopes, columns=['simulated_slope'])

# Plot the sampling distribution of the fitted slope coefficients
plt.figure(figsize=(10, 6))
sns.histplot(simulated_slopes_df['simulated_slope'], bins=30, kde=True)
plt.title('Sampling Distribution of Fitted Slope Coefficient under Null Hypothesis')
plt.xlabel('Simulated Slope Coefficient')
plt.ylabel('Frequency')
plt.axvline(x=0, color='red', linestyle='--', label='Null Hypothesis (slope = 0)')
plt.legend()
plt.show()

# Step 3: Calculate the 95% bootstrapped confidence interval for the slope
lower_bound = np.percentile(slope_coefficients, 2.5)
upper_bound = np.percentile(slope_coefficients, 97.5)

print(f"95% Bootstrapped Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
contains_zero = lower_bound <= 0 <= upper_bound
print(f"Does the 95% confidence interval contain 0? {'Yes' if contains_zero else 'No'}")

# Step 4: Compare simulated p-value with OLS regression
ols_model = smf.ols('duration ~ waiting', data=long_wait_data).fit()
p_value = ols_model.pvalues['waiting']
print(f"Simulated p-value from OLS model: {p_value:.4f}")

# Calculate simulated p-value
simulated_p_value = np.mean(np.array(simulated_slopes) >= 0)  # Assuming one-sided test for null hypothesis
print(f"Simulated p-value under null hypothesis: {simulated_p_value:.4f}")
```

Sampling Approaches
1. Permutation Test:
    - This method involves shuffling group labels (e.g., "short" vs. "long") to generate a distribution of mean differences under the null hypothesis. It assesses whether observed differences are significant by comparing the actual mean difference to this distribution.

2. Bootstrap Confidence Interval:
    - Bootstrapping creates resampled datasets by sampling with replacement. For each resample, we calculate the mean difference between groups and construct a 95% confidence interval using the 2.5th and 97.5th percentiles of these differences, providing a range for the mean difference's uncertainty.

Comparison with Indicator Variable-Based Model
- Objective:
    - All three methods assess mean differences between groups. The permutation and bootstrap methods are non-parametric and do not assume a specific data distribution, while the indicator variable model assumes a linear relationship.
- Output:
    - The permutation test yields a p-value indicating significance, while the bootstrap method provides a confidence interval. The indicator model produces estimated coefficients for predictors.
- Flexibility:
    - Permutation and bootstrap methods are versatile for various statistics and suitable for non-normal data. The indicator variable model is structured but less flexible, making it appropriate for linear relationships. Choosing between them depends on data characteristics and research questions.
